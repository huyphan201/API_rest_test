```python
def get_all(query):
    conn = sqlite3.connect("data/newsdb")
    dt = conn.execute(query).fetchall()
    conn.close()

    return dt

def get_url_id():
    ur = get_all("SELECT * FROM urlnews")
    return ur

    
def add_news(conn, dom, url,url_id):
    # sql = """
    # INSERT INTO news(subject, description, image, original_url, category_id)
    # VALUES (?, ?, ?, ?, ?)
    # """

    sql = """
    INSERT INTO news(domain, title, time, description, image, content, author, url_id)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    """

    article = Article(url)
    article.download()
    article.parse()

    # conn.execute(sql, (article.title ,article.text, article.top_image,
    #                                 article.url,category_id))
    #conn.commit()



    title = article.title
    time = article.publish_date
    descrip = "".join(article.text.split("\n")[0])
    img  = article.top_image
    content = " ".join(article.text.split("\n")[1:])
    authors = "".join(article.authors)
    #ags = ",".join(article.tags)
    #meta = article.meta_data

    #print(tags)
    conn.execute(sql, (dom, title, time, descrip, img,
                                      content, authors, url_id))
    conn.commit()


def get_url_crawled(conn,id_dom):
    fi  = in_out.json_input()
    sql = """
            INSERT INTO urlnews(names, urlsnews, id_dom)
            VALUES (?, ?, ?)
            """
    for f in fi:
        # try:
       
        old = sosa_file.url_up()
        for o in old:
            #print(o)
            #conn.execute(sql, (f["web_name"], o, id_dom))
            #conn.commit()
            id_u = get_url_id()
            for i in id_u:
                add_news(conn, f["web_name"], o,i[0])

        # except AttributeError:
        #     pass
        
                
       

def get_news_url():
    conn = sqlite3.connect("/home/norii/NORII/test/data/newsdb")
    cats = get_all("SELECT * FROM domain")
    
    for cat in cats:
        cat_id = cat[0]
        #print(cat_id)
        #try:
            #print("+++", article.url)
        get_url_crawled(conn,cat_id)
            #add_news(conn, article.url, cat_id)
        # except Exception as ex:
        #     print('Error' + str(ex))
        #     pass


    conn.close()

if __name__=="__main__":
    get_news_url()
    
    #print(get_all("select * from category"))
```
